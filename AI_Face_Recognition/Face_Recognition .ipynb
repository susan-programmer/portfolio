```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow  # مخصوص نمایش در Colab

# نام تصویر آپلود شده را وارد کنید
image_path = "6.jfif"  # جایگزین با نام فایل

# بارگذاری مدل تشخیص چهره Haar Cascade
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# خواندن تصویر
image = cv2.imread(image_path)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# تشخیص چهره‌ها
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5, minSize=(30, 30))

# رسم مستطیل دور چهره‌های شناسایی‌شده
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 3)

# نمایش تصویر خروجی در Google Colab
cv2_imshow(image)
```
